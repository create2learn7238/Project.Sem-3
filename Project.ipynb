```python
##### from datetime import datetime
import random
import string
import os

class Login:
    doctor_dict = {"doc1": "123"}
    def doctor_login(self, username):
        self.username = username
        print("Welcome :", self.username)

        while True:
            print("\n 1. View Patient \n 2. Log out ")
            doc_choice = int(input("Doctor Enter Your Choice: "))

            if doc_choice == 1:
                print("Appoint methode me se ek patient out ")
            elif doc_choice == 2:
                print("Logging OUT ...")
                break
            else:
                print("Invalid Choice")

    def admin(self):
        print("Welcome Admin")
        while True:
            print("\n 1. Add Patient \n 2. Add Doctor \n 3. Add Appointment \n 4. Logout")

            try:
                choice = int(input("Enter Your Choice: "))
            except ValueError:
                print("Please enter a valid number")
                continue

            if choice == 1:
                print("===== Patient details =====")
                firstName = input("Enter FirstName: ")
                lastName = input("Enter lastName: ")
                name = firstName + " " + lastName
                date = datetime.now()
                age = int(input("Enter Age: "))
                mobile = int(input("Enter Mobile Number: "))
                pat_id = chr(random.randint(65,90))+chr(random.randint(65,90)) + str(random.randint(0,99))

                gender = "Female" if int(
                    input("\n 1. Female \n 2. Male \n Choose: ")
                ) == 1 else "Male"

                with open(firstName + ".txt", "w") as file:
                    file.writelines([
                        f"Name: {name}\n",
                        f"Patient Id: {pat_id}\n"
                        f"Date: {date}\n",
                        f"Age: {age}\n",
                        f"Gender: {gender}\n",
                        f"Mobile: {mobile}\n"
                    ])

                file_exists=os.path.exists("allPatient.csv")
                with open("allPatient.csv", "a") as alldetails:
                    if not file_exists :
                        alldetails.write("Name,Age,Gender,Mobile,Patient_id\n")
                    alldetails.write(
                        name + "," + str(age) + "," + gender + "," + str(mobile) + str(pat_id)+"\n"
                    )

                
                print("Patient Added Successfully")
                print("Patient ID generated:", pat_id)
                print(f"Patient record saved to {firstName}.txt")

            elif choice == 2:
                print("===== Doctor Details =====")
                doc_id = chr(random.randint(65,90))+chr(random.randint(65,90)) + str(random.randint(0,99))

                doc_name = input("Enter Doctor Name: ")
                specialization = input("Enter Specialization: ")

                file_exists= os.path.exists("doctors.csv")
                with open("doctors.csv", "a") as file:
                    if not file_exists:
                        file.write("DoctorName,Specialization,DoctorID\n")
                    file.write(doc_name + "," + specialization + "," + doc_id + "\n")

                print("Doctor Added Successfully")
                print("Doctor ID generated:", doc_id)
                Login.doctor_dict[doc_name] = doc_id

            elif choice == 4:
                break

    def menu(self):
        patient_dict = {"pat1": "p@123"}

        while True:
            print("\n 1. Reception \n 2. Doctor \n 3. Patient \n 4. Chemist \n 5. Exit")

            try:
                choice = int(input("Enter Your Choice: "))
            except ValueError:
                print("Please enter a valid number")
                continue

            if 0 < choice < 5:
                username = input("Enter Your User: ")

            if choice == 1:
                if username == "Admin":
                    pwd = input("Enter Your Pass: ")
                    if pwd == "Admin@123":
                        self.admin()
                    else:
                        print("Wrong PASS")
                else:
                    print("Invalid USER")

            elif choice == 2:
                if username in Login.doctor_dict:
                    pwd = input("Enter Your Pass: ")
                    if Login.doctor_dict[username] == pwd:
                        self.doctor_login(username)
                    else:
                        print("Incorrect password")
                else:
                    print("User Not found as :", username)

            elif choice == 3:
                if username in patient_dict:
                    pwd = input("Enter Your Pass: ")
                    if pwd == patient_dict[username]:
                        print("Going to patient method")
                    else:
                        print("Wrong PASS")
                else:
                    print("Invalid USER")
                break

            elif choice == 4:
                print("Chemist module")
                break

            elif choice == 5:
                print("Exiting >>>")
                break

            else:
                print("Enter Valid Choice")


obj = Login()
obj.menu()
```
